# IMDB Scraping
### Extracting movies relating to pandemics,  with year and rating

```python
#Imports
from splinter import Browser
from bs4 import BeautifulSoup
import re
```

```python
#Set up Splinter
executable_path = {'executable_path':'C:/paths/chromedriver.exe'}
browser = Browser('chrome', **executable_path)
```

```python
#Visit IMDB Page
url = 'https://www.imdb.com/search/title/?title_type=feature,tv_episode&plot=pandemic&view=simple&count=250'
browser.visit(url)
```

```python
#TODO Fix tiles containing the word 'Next'
all_entries = []
while soup.find(class_='next-page'):
    html = browser.html
    soup = BeautifulSoup(html, 'html.parser')
    entries = soup.find_all(class_='lister-col-wrapper')
    for entry in entries:
        info = {}

        title = entry.find('a').text
        info['title'] = title

        link = entry.find('a')['href']
        info['link'] = link

        try:
            year = entry.find(class_='lister-item-year').text
            year = re.search('[0-9]{4}', year).group()
            info['year'] = year
        except:
            info['year'] = ''
    
        try:
            rating = entry.find(class_='col-imdb-rating').find().text
            rating = re.search('\d\.\d|\d', rating).group()
            info['rating'] = rating
        except:
            info['rating'] = ''
        print(info)
        all_entries.append(info)
    soup.find(class_='next-page').click()
    print('Next Page')
```

```python
#Save JSON
import json
with open('/outputs/IMDB_Data.json', 'w', encoding='utf-8') as f:
    json.dump(all_entries, f, ensure_ascii=False, indent=4)
```
